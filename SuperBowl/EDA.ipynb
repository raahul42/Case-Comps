## Dealing with missing values

```python
(df.isnull().sum().sort_values(ascending=False)/4589.13).iloc[38:45]
```

```python
null_cols = df.isnull().sum().sort_values(ascending=False).head(39).index
```

```python
df_94 = df.drop(null_cols,axis=1)
```

```python
df_94.info()
```

```python
(df_94.isnull().sum().sort_values(ascending=False)/4589.13).head()
```

```python
label = pd.read_csv('./data/train_y.csv')
label.head()
```

```python
df_94 = df_94.join(label)
df_94.head()
```

```python
df_94.dropna().count()[0]/4589.13
```

```python
df_full = df_94.dropna()
```

## Exploratory Analysis

```python
df_full.info()
```

```python
'Default_Flag' in df_full.columns
```

```python
df_full['Default_Flag'].value_counts()
```

```python
label_0 = df_full[df_full['Default_Flag']==0]
label_x = df_full[df_full['Default_Flag'] != 0]
```

```python
label_x.describe() - label_0.describe()
```

```python
column_dict = pd.read_excel('Variable_Dictionary.xlsx')
column_dict.head()
```

```python
spend_variables = []

for i in range(column_dict['Variable Name'].size):
    if ( 'Spend' in column_dict['Description'].iloc[i]):
        spend_variables.append(column_dict['Variable Name'].iloc[i])

print(spend_variables)
```

```python
dropped_cols = ['D_73', 'S_7', 'S_19', 'S_23', 'S_25']

for i in dropped_cols:
    spend_variables.remove(i)
```

```python
label_x[spend_variables].describe() - label_0[spend_variables].describe()
```
